```python
def check_observed_GPP_exists(PLUMBER2_flux_path, varname, site_name):

    # Set input file path
    file_path    = glob.glob(PLUMBER2_flux_path +"/*"+site_name+"*.nc")
    # print(file_path)
    var_exist    = False
    var_name_in_model = []
    try:
        with nc.Dataset(file_path[0], 'r') as dataset:
            for var_name in dataset.variables:
                if varname.lower() == var_name.lower():
                    var_name_in_model.append(var_name)
                    var_exist         = True
                elif varname.lower() in var_name.lower():
                    var_name_in_model.append(var_name)
    except Exception as e:
        print(f"An error occurred: {e}, {site_name}, {file_path}")

    return var_name_in_model
```

```python
varname      = 'GPP'
# key_word     = 'gross'
# key_word_not = None

for site_name in site_names:
    GPP_lists= check_observed_GPP_exists(PLUMBER2_flux_path, varname, site_name)
    print(site_name, GPP_lists)
```
